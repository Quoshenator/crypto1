## 1.1Теорема Эйлера
Итак, освежим немного материал прошлого занятия, изучая арифметики остатков по модулю $Z_n$, мы с вами быстро наталкиваемся на примечательный факт,  если модуль не является простым числом, то в таблице умножения появляются результаты вида a * b = 0 mod n, например 2*3 =0, по модулю 6.

a*y = 1 mod n, не имеет однознаного решения если a имеет общие делители с n.

при том что a!=0 и b!=0, такие a и b называются делителями нуля, и в таких числовых системах появляются проблемы неоднознаного умножения, из-за которых у нас начинаются проблемы и со степенями таких чисел в чём легко убедиться на примере...

Но при этом можно заметить, что если число u и модуль n взаимно просты, то степень u и сама является циклической подгруппой как бы рассекая её. 



```python
import sys
import numpy as np
import PIL
%matplotlib inline 

n=10
squareMult=np.zeros((n,n),dtype="uint8")

np.set_printoptions(threshold=sys.maxsize,linewidth=150)
for i in range(n):
    for j in range(n):
        squareMult[i,j]=(i*j)%n
np.set_printoptions(threshold=None)
np.set_printoptions(edgeitems=3,infstr='inf',
linewidth=75, nanstr='nan', precision=8,
suppress=False, threshold=1000, formatter=None)
print(squareMult)
from matplotlib import pyplot as plt
fig=plt.figure(figsize=(18, 16), dpi= 80,)
plt.imshow(squareMult, interpolation='nearest')
plt.show()
```

```python
powN=np.zeros((n,n),dtype="uint8")
for i in range(n):
    for j in range(n):
        powN[i,j]=pow(i,j,n)
np.set_printoptions(threshold=sys.maxsize,linewidth=150)
print(powN)
fig=plt.figure(figsize=(18, 16), dpi= 80,)
plt.imshow(powN, interpolation='nearest')
plt.show()
```

Теперь перейдём собственно к использованию этих свойств в системе RSA.
Возьмём два больших простых числа p и q

умножив их получим число n, по свойству функции Эйлера её значение от  $\phi(n)= \phi(p)*\phi(q)=(p-1)(q-1)$n  будет равно (p-1)(q-1),
теперь найдём число e меньшее $e<\phi(n)$ и взаимно простое с ним.

```python
import random
p,q=17,19
n=p*q
phi=(p-1)*(q-1)
print(n,"длина")
print(phi, "эйлер")
```
